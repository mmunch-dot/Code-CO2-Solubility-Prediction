```python
from sklearn.utils import shuffle
import pandas as pd

df = pd.read_csv('all_data.csv', sep=";")

# Drop unite
df = df.drop([0])

# Convert mol -> mmol
for i in df.index:
    df.loc[i, "solubility"] = "{:.1f}".format(float(df.loc[i, "solubility"])*1e3)
    
# We consider missing salt values as 0
df = df.fillna(0)

# We transform all values as numbers
for att in ["temperature", "solubility", "fat", "protein", "salt", "water"]:
    df[att] = df[att].apply(pd.to_numeric)

# When a product has been tested multiple times under the same conditions
# We take the mean value
d_values = dict()
for i in df.index:
    key = tuple(df.loc[i].loc[["Type",
                             "temperature",
                             "fat", "protein", "salt",
                             "water"]].to_list() )
    
    if key not in d_values:
        d_values[key] = {"values":list(), "index":list()}
    
    d_values[key]["values"].append( df.loc[i, "solubility"] )
    d_values[key]["index"].append( i )

for key in d_values:
    if len(d_values[key]["index"]) > 1:
        
        mean = sum(d_values[key]["values"])/len(d_values[key]["values"])
        df = df.drop(d_values[key]["index"][1:])
        df.loc[ d_values[key]["index"][0], "solubility" ] = mean
```

```python
from sklearn.utils import shuffle

df = shuffle(df)
df = df.set_axis(range(157)) # reorganise the index

df.head()
```

### Learning

```python
from sklearn.ensemble import RandomForestRegressor
import shap
import matplotlib.pyplot as plt

X = df[ ["temperature", "fat", "protein", "water", "salt"] ]
y = df["solubility"]

clf = RandomForestRegressor()
clf.fit(X,y)
```

### SHAP

```python
# We remove temperature too high (which correspond to oil)
df = df.drop(df[df.temperature > 30].index)
X = df[ ["temperature", "fat", "protein", "water", "salt"] ]
y = df["solubility"]
```

```python
explainer = shap.Explainer(clf)
shap_values = explainer(X) # List of SHAP-values for each x in X
```

```python
shap.plots.bar(explainer(X))
```

```python
shap.plots.beeswarm(explainer(X))
```

```python
shap_interaction_values = explainer.shap_interaction_values(X) # List of matrix
```

```python
def shapInfluence(DF, inter, x_var, y_var, show_original = False):
    """
    Displays the evolution of the SHAP values along x_var and y_var
    """
    
    l_att = ["temperature", "fat", "protein", "water", "salt"]
    
    X = DF[x_var]
    y = DF[y_var]
    shap = [ i[l_att.index(x_var)][l_att.index(y_var)] for i in inter]

    plt.scatter(X,y, c= shap, cmap = 'viridis', s = 20)
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    
    if show_original:
        X = df[x_var]
        y = df[y_var]
        shap = [ i[l_att.index(x_var)][l_att.index(y_var)] for i in shap_interaction_values]

        plt.scatter(X,y, c= shap, cmap = 'viridis', edgecolors = 'red', s = 20)
    
    plt.colorbar(label="SHAP-value")
```

#### 1.1 Water/Temperature interaction

```python
shapInfluence(df, shap_interaction_values, "water", "temperature")
```

```python
shapInfluence(df, shap_interaction_values, "fat", "temperature")
```

```python
shapInfluence(df, shap_interaction_values, "protein", "temperature")
```

```python
shapInfluence(df, shap_interaction_values, "salt", "temperature")
```

### Extrapolation from a generated dataset

```python
from itertools import permutations
import random as rd

data = {} 

i = 0
for t in range(0,30):
    for att in permutations(["fat", "protein", "water", "salt"], 4):
        for seuil in [ [0,25], [25,50], [50,75], [75,100] ]:
            for ligne in range(5):

                # We fix a threshold for an attribute
                val = rd.randint(seuil[0], seuil[1])

                # We randomly generate the others
                k = [rd.random() for r in range(3)]
                rep = [val] + [(k[i]/sum(k))*(100-val) for i in range(3)]

                data[i] = {att[l]:rep[l] for l in range(4)}
                data[i]["temperature"] = t

                i += 1
            

X_exp = pd.DataFrame.from_dict(data, orient='index', columns = ["temperature", "fat", "protein", "water", "salt"])
```

```python
shap_interaction_values_exp = explainer.shap_interaction_values(X_exp) # List of matrix
```

```python
shapInfluence(X_exp, shap_interaction_values_exp, "water", "temperature", show_original = True)
```
