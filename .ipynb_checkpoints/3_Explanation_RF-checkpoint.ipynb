### Preparation

```python
import numpy as np
import pandas as pd
import os

df = pd.read_csv('all_data.csv', sep=";")

# Drop
df = df.drop([0]) # Unités

# Convert mol -> mmol
for i in df.index:
    df.loc[i, "solubility"] = "{:.1f}".format(float(df.loc[i, "solubility"])*1e3)
    
# Fill na
df = df.fillna(0)

for att in ["temperature", "solubility", "fat", "protein", "salt", "water"]:
    df[att] = df[att].apply(pd.to_numeric)
    
# Outliers
df = df.drop(df[df.solubility < 6].index)
df = df.drop(df[df.solubility > 70].index)
df = df.drop(df[df["product"] == "Snellina"].index)
```

```python
# Prend la valeur moyenne quand des lignes se répètent
d_values = dict()
for i in df.index:
    key = tuple(df.loc[i].loc[["Type",
                             "temperature",
                             "fat", "protein", "salt",
                             "water"]].to_list() )
    
    if key not in d_values:
        d_values[key] = {"values":list(), "index":list()}
    
    d_values[key]["values"].append( df.loc[i, "solubility"] )
    d_values[key]["index"].append( i )

for key in d_values:
    if len(d_values[key]["index"]) > 1:
        
        mean = sum(d_values[key]["values"])/len(d_values[key]["values"])
        df = df.drop(d_values[key]["index"][1:])
        df.loc[ d_values[key]["index"][0], "solubility" ] = mean
```

```python
from sklearn.utils import shuffle

df = shuffle(df)
df = df.set_axis(range(137))

df.shape
```

### Apprentissage

```python
from sklearn.ensemble import RandomForestRegressor
import shap

X = df[ ["temperature", "fat", "protein", "water", "salt"] ]
y = df["solubility"]

clf = RandomForestRegressor()
clf.fit(X,y)
```

### Explication

```python
explainer = shap.Explainer(clf)
shap_values = explainer(X)
```

### Etude d'un jeu genere

```python
from itertools import permutations
import random as rd

data = {} 

i = 0
for t in range(0,26,5):
    for att in permutations(["fat", "protein", "water", "salt"], 4):
        for seuil in [ [0,25], [25,50], [50,75], [75,100] ]:
            for ligne in range(10):

                # On fixe un seuil pour un attribut
                val = rd.randint(seuil[0], seuil[1])

                # On genere les autres attributs
                k = [rd.random() for r in range(3)]
                rep = [val] + [(k[i]/sum(k))*(100-val) for i in range(3)]

                data[i] = {att[l]:rep[l] for l in range(4)}
                data[i]["temperature"] = t

                i += 1
            

X_exp = pd.DataFrame.from_dict(data, orient='index', columns = ["temperature", "fat", "protein", "water", "salt"])
```

### Generation d'un jeu "similaire"

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
from math import isnan
from sklearn.preprocessing import KBinsDiscretizer

def detNbCat(X):
    sse = dict()
    sil = dict()
    for k in range(1,9):
        kmeans = KMeans(n_clusters=k).fit(X)
        sse[k] = kmeans.inertia_

        if k>1:
            sil[k] = silhouette_score(X, kmeans.labels_, metric='euclidean')

    plt.figure()
    plt.plot(list(sse.keys()), list(sse.values()))
    plt.xlabel("Number of cluster")
    plt.ylabel("SSE")
    plt.title("Elbow Method")
    plt.show()

    plt.figure()
    plt.plot(list(sil.keys()), list(sil.values()))
    plt.xlabel("Number of cluster")
    plt.ylabel("Silhouette Coefficient")
    plt.title("Silhouette Coefficient Method")
    plt.show()
```

```python
x = list()
atts = ["fat", "protein", "salt", "water"]

# Base
for i in df.index:
    x.append( list(df.loc[i, atts]) )

new_x = list()
for l in x:
    if not any( [isnan(k) for k in l] ):
        #print(l, any( [isnan(k) for k in l] ))
        new_x.append(l)
        
print("{} / {}".format(len(new_x), len(x)))

detNbCat(new_x)
```

```python
K = 4

kmeans =  KMeans(n_clusters=K).fit(new_x)
profiles = pd.DataFrame(kmeans.cluster_centers_, columns = atts)
profiles
```

```python
import numpy as np

dc = {i:{"fat":[], "protein":[], "salt":[], "water":[]} for i in range(K)}
pred = kmeans.predict(new_x)

for i in range(len(pred)):
    dc[pred[i]]["fat"].append(new_x[i][0])
    dc[pred[i]]["protein"].append(new_x[i][1])
    dc[pred[i]]["salt"].append(new_x[i][2])
    dc[pred[i]]["water"].append(new_x[i][3])

# Genere des items similaires
data = {}
it = 1

for i in dc:
    param = {"fat":{}, "protein":{}, "salt":{}, "water":{}}
    for j in ["fat", "protein", "salt", "water"]:
        param[j]["mu"] = np.mean(dc[i][j])
        param[j]["sigma"] = np.var(dc[i][j])
    
    
    for t in range(0,26,5):
        for j in range(100):
            data[it] = {"temperature":t}
            
            for att in ["fat", "protein", "salt", "water"]:
                val = np.random.normal(param[att]["mu"], param[att]["sigma"])
                data[it][att] = min(100, max(0, val )  )
                
            it += 1
            
X_sim = pd.DataFrame.from_dict(data, orient='index', columns = ["temperature", "fat", "protein", "water", "salt"])        
```

```python
shap_interaction_values = shap.TreeExplainer(clf).shap_interaction_values(X_sim)
shap.summary_plot(shap_interaction_values, X_sim)
```
